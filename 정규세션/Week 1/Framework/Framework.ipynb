### 1. Library와 Framework의 차이 간단하게 서술하기. (100자 내외)

##### Library
프로그래밍 맥락에서 라이브러리는 본질적으로 사용자가 자신의 코드에서 특정 기능을 구현하기 위해 호출할 수 있는 미리 작성된 코드 모음    
필요에 따라 사용하도록 선택할 수 있는 도구 세트 또는 기능 세트   
예를 들어, Python에서 작업하는 경우 수치 계산에는 numpy 라이브러리를 사용하고 데이터 플로팅에는 matplotlib를 사용    
라이브러리의 주요 측면은 프로그래머가 제어할 수 있다는 것이며 언제 라이브러리의 함수를 호출할지와 그 결과로 무엇을 할지 결정    

##### Framework
프레임워크는 도구뿐만 아니라 애플리케이션 개발 방법과 코드 구성 방법에 대한 구조도 제공    
본질적으로 프레임워크는 프로젝트의 아키텍처를 결정   
자체 디자인 패턴과 규칙 세트가 제공되며 프로그래머는 자신의 코드를 이 구조에 맞춰야 함    
예를 들어 Python 웹 개발을 위한 Django 및 Flask가 있으며, 여기서 프레임워크는 웹 애플리케이션을 구조화하고, 요청을 처리하고, 데이터베이스와 상호 작용하는 방법을 간략하게 설명

### 2. Deep Learning과 Machine Learning의 차이 간단하게 서술하기

##### Machine Learning
머신러닝은 데이터로부터 학습하고 이를 기반으로 의사결정을 내릴 수 있는 시스템을 구축하는 데 중점을 두는 인공지능 분야    
ML은 명시적인 프로그래밍이 불가능한 작업을 수행하도록 컴퓨터를 가르치는 다양한 기술을 포함    
예시로는 선형 회귀, 의사결정 트리, 지원 벡터 머신과 같은 알고리즘이 포함    
ML의 학습 프로세스는 결정을 내리는 데 복잡한 구조가 필요하지 않으며 다양한 데이터 유형을 처리할 수 있음    

##### Deep Learning
딥 러닝은 특히 여러 계층이 있는 신경망을 사용하여 데이터의 복잡한 패턴을 모델링하는 기계 학습의 하위 집합    
이는 모델이 명시적인 특징 추출 없이 원시 데이터에서 직접 특징을 식별하는 방법을 학습하는 이미지 및 음성 인식과 같은 작업에 특화    
딥 러닝이 제대로 작동하려면 상당한 양의 데이터가 필요하고 계산 집약적이지만 자연어 처리, 이미지 인식 등과 같은 작업에서는 좋은 성능을 보임
