### 결측치 & 이상치 확인

```python
# 결측치 확인

missing_values = data.isnull().sum()
missing_values[missing_values > 0]

## 확인결과, 결측치는 존재하지 않는 것으로 확인
```

```python
# 이상치 확인

# 이상치 확인을 위해 수치형 변수들만을 저장
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# IQR 기준으로 이상치가 존재하는지 한눈에 확인하기 위해 BoxPlot 확인
plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_columns):
    plt.subplot(5, 7, i + 1)
    sns.boxplot(data[col])
    plt.title(col)

plt.tight_layout()
plt.show()

# BoxPlot을 확인해본 결과, MonthlyIncome, NumCompaniesWorked, PerformanceRating, StockOptionLevel, TotalWorkingYears, TrainingTimesLastYear, 
# YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion, YearsWithCurrManager 에서 이상치가 확인
```

```python
# IQR 기준 이상치라고 판단되는 값을 모두 중앙값으로 대체하는 함수
def replace_outliers_with_median(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    median = df[column].median()
    df[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
    return df

# 해당 함수를 활용해 이상치들을 중앙값으로 대체
for col in numerical_columns:
    data = replace_outliers_with_median(data, col)
```

```python
# 이상치 대체
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(20, 15))
for i, col in enumerate(numerical_columns):
    plt.subplot(5, 7, i + 1)
    sns.boxplot(data[col])
    plt.title(col)

plt.tight_layout()
plt.show()

# 이상치를 처리한 결과 이전보다 확연히 이상치가 줄어든 모습을 확인할 수 있다
```

### 데이터 시각화

```python
sns.set(style="whitegrid")

# Plot 1: 이직 여부에 따른 연령 분포 (Attrition by Age)
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='Age', hue='Attrition', shade=True)
plt.title('Attrition by Age (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

## 해당 그래프를 확인해본 결과, 3-40 대에서 이직이 가장 많이 존재하고, 나이가 많아질수록 이직률이 떨어지는 것을 확인
```

```python
# Plot 2: 부서별 직원 분포 (Employee Distribution by Department)
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Department')
plt.title('Employee Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()

## 부서별로 직원 분포를 확인해본 결과, Research & Development 에 가장 많은 사람이 분포
```

```python
# Plot 3: 교육 수준별 이직률 (Attrition by Education Level)
turnover_rate = data.groupby('Education')['Attrition'].apply(lambda x: x[x == 'Yes'].count() / x.count())

plt.figure(figsize=(10, 6))
turnover_rate.plot(kind='bar')
plt.title('Turnover Rate by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Turnover Rate')
plt.xticks(rotation=0)
plt.show()

## 교육 수준이 높아질수록 이직률이 낮아지는 경향을 보임
```

```python
# Plot 4: 연간 소득별 직원 분포 (Employee Distribution by Annual Income)
plt.figure(figsize=(14, 6))
sns.histplot(data=data, x='MonthlyIncome', bins=20, kde=True)
plt.title('Employee Distribution by Annual Income')
plt.xlabel('Monthly Income')
plt.ylabel('Number of Employees')
plt.show()

## 연간 소득에 따라 직원 수가 어떻게 분포하는지를 확인해봤을 때 특이하게 정규분표 형태가 아닌 쌍봉분포가 나타남
```

```python
# Plot 5: 업무 만족도별 이직률 (Attrition by Job Satisfaction)
turnover_rate = data.groupby('JobSatisfaction')['Attrition'].apply(lambda x: (x=='Yes').sum() / len(x))

turnover_rate_df = turnover_rate.reset_index()
turnover_rate_df.columns = ['JobSatisfaction', 'TurnoverRate']

plt.figure(figsize=(10, 6))
plt.bar(turnover_rate_df['JobSatisfaction'], turnover_rate_df['TurnoverRate'])
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Turnover Rate')
plt.title('Turnover Rate by Job Satisfaction')
plt.xticks(turnover_rate_df['JobSatisfaction'])
plt.show()

## 만족도가 높을수록 이직률이 감소하는 모습을 보임
```

### 파생변수 생성

```python
# Create derived variables

# 1. AgeGroup
bins = [18, 30, 40, 50, 60, 70]
labels = ['20-30', '31-40', '41-50', '51-60', '61-70']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

## 연령대 별 그룹화 진행

# 2. TotalYearsAtCompany
data['TotalYearsAtCompany'] = data['YearsAtCompany'] + data['YearsInCurrentRole'] + \
                              data['YearsSinceLastPromotion'] + data['YearsWithCurrManager']

## 회사와 관련된 모든 연수에 대해 종합한 변수 생성

# 3. WorkLifeBalanceScore
data['WorkLifeBalanceScore'] = (data['WorkLifeBalance'] + data['JobSatisfaction'] + \
                                data['RelationshipSatisfaction']) / 3

## 워라밸 + 직업 만족도 + 관계 만족도에 대해 종합적인 점수를 파악하는 변수 생성

# 4. MonthlyIncome_Average
average_wage_by_education = data.groupby('Education')['MonthlyIncome'].mean()
data = data.join(average_wage_by_education, on='Education', rsuffix='_Average')

## 교육 수준 별 평균 임금에 대한 변수 생성

# 5. avg_work_life_balance
avg_work_life_balance = data.groupby('AgeGroup')['WorkLifeBalanceScore'].mean().reset_index()
avg_work_life_balance = avg_work_life_balance.rename(columns={'WorkLifeBalanceScore': 'avg_work_life_balance'})
data = data.merge(avg_work_life_balance, on='AgeGroup')

## 나이대 별 워라밸 평균에 대한 변수 생성
```
